```python
def get_page_nums(row,textbook):
    slc_concepts = [x.lower() for x in set(row['concept'].split('_'))] if not(pd.isna(row['concept'])) else []
    ind = 0
    page_nums = {}

    matched_concepts = {}
    while ind < textbook.shape[0]:
        row1 = textbook.loc[ind]
        book_concepts = [x.lower() for x in set(row1['concept'].split('_'))]
        slc_match_count = 0
        book_match_count = 0

        for concept in book_concepts:
            if concept in slc_concepts: slc_match_count += 1
        
        for concept in slc_concepts:
            if concept in book_concepts: book_match_count += 1
        
        if slc_match_count == book_match_count and len(slc_concepts) == slc_match_count:
            page_nums[f"{row1['section_id']}-{row1['page_num']}"] = slc_match_count #slc_match_count if slc_match_count > book_match_count else 0
            matched_concepts[f"{row1['section_id']}-{row1['page_num']}"] = row1['concept']

        ind += 1
    
    get_max_count = max(page_nums.values()) if len(page_nums.values()) > 0 else 0

    return min([(k,v,matched_concepts[k]) for k,v in page_nums.items() if v == get_max_count]) if get_max_count > 0 else ('','','')


content_name_urls['book_match'] = content_name_urls.apply(get_page_nums,1,args=(textbook,))
# content_name_urls = content_name_urls.explode('book_match')
content_name_urls['match_count'] = content_name_urls.apply(lambda x: x['book_match'][1],axis=1)
content_name_urls['book_concepts'] = content_name_urls.apply(lambda x: x['book_match'][2],axis=1)
content_name_urls['page_numbers'] = content_name_urls.apply(lambda x: x['book_match'][0],axis=1)
content_name_urls[['content_name','content_type','display_name','url','page_numbers','book_concepts','concept']]   #,'section_ids']]
```
