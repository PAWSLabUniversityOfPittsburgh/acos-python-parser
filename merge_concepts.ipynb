```python
textbook = pd.read_csv('smart_learning_content_section_page_numbers.csv')

textbook = textbook[['section_id','page_id','concept']]

textbook.loc[:,'section_id'] = textbook.loc[:,'section_id'].apply(lambda x: f"Section {'.'.join(x.split('-')[1:])}")

textbook = textbook.rename(columns={
    "page_id":"page_num"
})

textbook.head()
```

```python
textbook = textbook.sort_values(by='page_num').reset_index()
textbook.loc[:,'concept_upto'] = textbook.loc[:,'concept']
textbook.to_csv('./out_merged_files/textbook_concepts_upto.csv')
textbook
```

```python
for ind, row in textbook.iterrows():
    if ind > 0:
        prev_concepts = []
        for j in range(ind):
            if not(pd.isna(textbook.loc[j,'concept'])):
                prev_concept = list(set(textbook.loc[j,'concept'].split('_')))
                prev_concepts.extend(prev_concept)
                # print(j,ind,textbook.loc[j,'concept'])
                # print(list(set(prev_concepts)))
        
        if not(pd.isna(row['concept'])):
            prev_concepts.extend(list(set(row['concept'].split('_'))))
            
        prev_concepts = list(set(prev_concepts))


        textbook.loc[ind,'concept_upto'] = '_'.join(prev_concepts)

textbook = textbook.rename(columns={
    "concept": "present_concept",
    "concept_upto": "concept"
})

textbook
```

```python
content_name_urls['page_numbers'] = [[] for x in range(len(content_name_urls.index))]

for ind1, row1 in content_name_urls.iterrows():

    slc_concept = [x.lower() for x in set(row1['concept'].split('_'))] if not(pd.isna(row1['concept'])) else []
    page_numbers = []
    section_ids = []
    prev_matched = False
    for ind2, row2 in textbook.iterrows():
        if not(prev_matched):
            count_match = 0
            for concept in slc_concept:
                book_concepts = [x.lower() for x in set(row2['concept'].split('_'))]
                if concept in book_concepts:
                    count_match+=1
            if count_match == len(slc_concept) and len(slc_concept) < len(book_concepts):
                # page_numbers.append(row2['page_num'])
                # section_ids.append(row2['section_id'])
                page_numbers.append(f"{row2['section_id']}-{row2['page_num']}")
                prev_matched=True

    content_name_urls.loc[ind1,'page_numbers'] = ','.join([str(x) for x in set(page_numbers)])
    # content_name_urls.loc[ind1,'section_ids'] = ','.join([str(x) for x in set(section_ids)])

content_name_urls[['content_name','content_type','display_name','url','page_numbers']]#,'section_ids']]
```

```python
content_name_urls.to_csv('./out_merged_files/content_name_concept_urls_page_numbers.csv',index=False)
```
